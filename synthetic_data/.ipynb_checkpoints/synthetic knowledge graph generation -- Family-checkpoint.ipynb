```python
import numpy as np
from copy import deepcopy
import torch
import random
```

```python
max_node_num = 400
# num_root = 10
p_parents = 0.8
p_couple = 1.0
p_child = 0.4
node_num=0
person_list = []
edge_list = []
max_generation = 3

class Person:
    def __init__(self, node=node_num, gender="female", generation=0):
        global node_num, person_list;
        self.name = node
        self.gender = gender
        self.father = None
        self.mother = None
        self.children = None
        self.husband = None
        self.wife = None
        self.age = None
        self.birthplace = None
        self.job = None
        self.generation = generation
        node_num += 1
        person_list.append(self)
        
    def build(self):
        global node_num, person_list;
        if node_num >= max_node_num:
            return
        new_person_list = []
        if self.father is None:
            if np.random.rand(1) < p_parents and self.generation > 0:
                self.father = Person(node=node_num, gender="male", generation=self.generation-1)
                self.mother = Person(node=node_num, gender="female", generation=self.generation-1)
                self.father.wife = self.mother
                self.mother.husband = self.father
#                 self.father.build()
                new_person_list.append(self.father)
                if self.father.children is None:
                    self.father.children = [self]
                else:
                    self.father.children.append(self)
                self.mother.children = self.father.children
#                 self.mother.build()
                new_person_list.append(self.mother)
                
        if self.gender == "female" and self.husband is None:
            if np.random.rand(1) < p_couple:
                self.husband = Person(node=node_num, gender="male", generation=self.generation)
                self.husband.wife = self
                if self.generation < max_generation:
                    if self.children is None:
                        self.children = []
                    num_children = min(np.random.geometric(p=p_child), 4)
                    if num_children > len(self.children):
                        for _ in range(num_children-len(self.children)):
                            if np.random.rand(1) < 0.5:
                                gender = "female"
                            else:
                                gender = "male"
                            child = Person(node=node_num, gender=gender, generation=self.generation+1)
                            child.mother = self #switch to father if male
                            child.father = self.husband
    #                         child.build()
                            new_person_list.append(child)
                            self.children.append(child)
                
                self.husband.children = self.children
                new_person_list.append(self.husband)
#                 self.husband.build()
        elif self.gender == "male" and self.wife is None:
            if np.random.rand(1) < p_couple:
                self.wife = Person(node=node_num, gender="female", generation=self.generation)
                self.wife.husband = self
                if self.generation < max_generation:
                    if self.children is None:
                        self.children = []
                    num_children = min(np.random.geometric(p=p_child), 4)
                    if num_children > len(self.children):
                        for _ in range(num_children-len(self.children)):
                            if np.random.rand(1) < 0.5:
                                gender = "female"
                            else:
                                gender = "male"
                            child = Person(node=node_num, gender=gender, generation=self.generation+1)
                            child.father = self #switch to father if male
                            child.mother = self.wife
    #                         child.build()
                            new_person_list.append(child)
                            self.children.append(child)
                self.wife.children = self.children
                new_person_list.append(self.wife)
#                 self.wife.build()
        for person in new_person_list:
            person.build()
                
```

```python
def build_graph(person):
    global tuple_list, if_build;
    if if_build[person.name] == 1:
        return
    if_build[person.name] = 1
        
    if person.father is not None:
        tuple_list.append((person.name, person.father.name, "father"))
        build_graph(person.father)
    if person.mother is not None:
        tuple_list.append((person.name, person.mother.name, "mother"))
        build_graph(person.mother)
    if person.children is not None:
        if len(person.children) > 1:
            for child_1 in person.children:
                for child_2 in person.children:
                    if child_1.name >= child_2.name:
                        continue
                    if child_2.gender == "male":
                        tuple_list.append((child_1.name, child_2.name, "brother"))
                        if child_1.children is not None:
                            for child_child_1 in child_1.children:
                                tuple_list.append((child_child_1.name, child_2.name, "uncle"))
                    else:
                        tuple_list.append((child_1.name, child_2.name, "sister"))
                        if child_1.children is not None:
                            for child_child_1 in child_1.children:
                                if (child_child_1.name, child_2.name) == (5, 3):
                                    print("?", child_1.name, child_2.name, child_child_1.name)
                                tuple_list.append((child_child_1.name, child_2.name, "aunt"))
                                    
                    if child_1.gender == "male":
                        tuple_list.append((child_2.name, child_1.name, "brother"))
                        if child_2.children is not None:
                            for child_child_2 in child_2.children:
                                tuple_list.append((child_child_2.name, child_1.name, "uncle"))
                    else:
                        tuple_list.append((child_2.name, child_1.name, "sister"))
                        if child_2.children is not None:
                            for child_child_2 in child_2.children:
                                tuple_list.append((child_child_2.name, child_1.name, "aunt"))
            
                    if child_1.children is not None:
                        for child_child_1 in child_1.children:
                            if child_child_1.gender == "male":
                                tuple_list.append((child_2.name, child_child_1.name, "nephew"))
                            else:
                                tuple_list.append((child_2.name, child_child_1.name, "niece"))
                                
                    if child_2.children is not None:
                        for child_child_2 in child_2.children:
                            if child_child_2.gender == "male":
                                tuple_list.append((child_1.name, child_child_2.name, "nephew"))
                            else:
                                tuple_list.append((child_1.name, child_child_2.name, "niece"))
                                
            
        for child in person.children:
            tuple_list.append((person.name, child.name, "child"))
            build_graph(child)
            
    if person.husband is not None:
        tuple_list.append((person.name, person.husband.name, "husband"))
        build_graph(person.husband)
        
    if person.wife is not None:
        tuple_list.append((person.name, person.wife.name, "wife"))
        build_graph(person.wife)
        
    
```

```python
#From random expanded objects to a graph with family relations
tuple_list = []
if_build = np.zeros([node_num+1])
build_graph(root_person)
print(len(tuple_list))
tuple_list = list(set(tuple_list))
edge_list = []
edge_type_list = []
for (person_1, person_2, edge_type) in tuple_list:
    edge_list.append((person_1, person_2))
    edge_type_list.append(edge_type)
print(len(edge_list))
```

```python
#Visualization
import matplotlib.pyplot as plt
import networkx as nx

edge_labels = {}
for edge, edge_type in zip(edge_list, edge_type_list):
    edge_labels[edge] = edge_type

edge_list_filter = []
edge_type_list_filter = []
edge_labels_filter = {}
for edge, edge_type in zip(edge_list, edge_type_list):
    if edge_type in ["father", "mother", "husband"]: #
#         edge_list_filter.append((edge[1], edge[0]))
        if edge_type == "husband":
            edge_type = "spouse"
            if np.random.rand(1) < 0.5:
                edge_list_filter.append((edge[0], edge[1]))
            else:
                edge_list_filter.append((edge[1], edge[0]))
        else:
            edge_list_filter.append((edge[0], edge[1]))
        edge_type_list_filter.append(edge_type)
        edge_labels_filter[edge] = edge_type

G = nx.DiGraph()
G.add_edges_from(edge_list_filter)
# pos = nx.spring_layout(G) #multipartite_layout
# pos = nx.multipartite_layout(G) #


for layer, nodes in enumerate(nx.topological_generations(G)):
    # `multipartite_layout` expects the layer as a node attribute, so add the
    # numeric layer value as a node attribute
    for node in nodes:
        G.nodes[node]["layer"] = layer

# Compute the multipartite_layout using the "layer" node attribute
# pos = nx.multipartite_layout(G, subset_key="layer")
pos = nx.spring_layout(G, k=5/np.sqrt(G.order()))

plt.figure(figsize=(10, 7))
nx.draw(
    G, pos, 
#     edge_color='black', width=1, linewidths=1,
#     G, edge_color='black', width=1, linewidths=1,
#     node_size=500, node_color='pink', alpha=0.9, arrowstyle="->",arrowsize=10,
    labels={node: node for node in G.nodes()}
)

nx.draw_networkx_edge_labels(
    G, pos,
    edge_labels=edge_labels_filter,
    font_color='red',
    font_size=10,
)
plt.axis('off')
plt.show()
```

```python
#BUILD NAMES
names = []
# f = open("random_female_full_names.csv", "r")
# female_name_list = [line.strip() for line in f.readlines()][1:]
# f = open("random_male_full_names.csv", "r")
# male_name_list = [line.strip() for line in f.readlines()][1:]
female_name_list = torch.load("random_female_full_names.pt")
male_name_list = torch.load("random_male_full_names.pt")
for person in person_list:
    if person.gender == "female":
        names.append(female_name_list.pop())
    else:
        names.append(male_name_list.pop())
        
fixed_names = deepcopy(names)

if_build = np.zeros([node_num+1])
def build_name(person):
    global edge_list, edge_type_list, if_build;
    if if_build[person.name] == 1:
        return
    if_build[person.name] = 1
        
    if person.father is not None:
        last_name = fixed_names[person.father.name].split(" ")[1]
        fixed_names[person.name] = fixed_names[person.name].split(" ")[0] + " " + last_name
        build_name(person.father)
    if person.mother is not None:
        build_name(person.mother)
    if person.children is not None:
        for child in person.children:
            build_name(child)
            
    if person.husband is not None:
        build_name(person.husband)
        if np.random.rand(1) < 0.2:
            last_name = fixed_names[person.husband.name].split(" ")[1]
            fixed_names[person.name] = fixed_names[person.name].split(" ")[0] + " " + last_name
        
    if person.wife is not None:
        build_name(person.wife)
        
np.random.seed(2)
build_name(root_person)

last_names = [name.split(" ")[1] for name in fixed_names]
last_name_dict = {}
for last_name in last_names:
    if last_name in last_name_dict.keys():
        last_name_dict[last_name] += 1
    else:
        last_name_dict[last_name] = 1

plt.figure(figsize=(15, 5))
plt.bar(last_name_dict.keys(), last_name_dict.values())
plt.xticks(range(len(last_name_dict.keys())), last_name_dict.keys(), rotation=45, fontsize=20)
plt.show()
```

```python
#Generate bio (birthday, birthplace, job)
#1. birthday
for person in person_list:
    person.age = None
def build_birthday(person, age):
    person.age = int(age)
        
    if person.mother is not None:
        if person.mother.age is None:
            mother_age_when_birth = 27 + np.clip(np.random.randn(1)* 3, -6, 6)[0]
            build_birthday(person.mother, age=age + mother_age_when_birth)
        
    if person.father is not None:
        if person.father.age is None:
            build_birthday(person.father)
            
    if person.husband is not None:
        if person.husband.age is None:
            husband_age = age + 2 + np.clip(np.random.randn(1)* 3, -6, 6)[0]
            build_birthday(person.husband, husband_age)
            
        if person.children is not None:
            for child in person.children:
                if child.age is None:
                    child_age = age - (27 + np.clip(np.random.randn(1)* 3, -6, 6))[0]
                    build_birthday(child, child_age)
        
    if person.wife is not None:
        if person.wife.age is None:
            wife_age = age + -2 + np.clip(np.random.randn(1)* 3, -6, 6)[0]
            build_birthday(person.wife, wife_age)

np.random.seed(3)
build_birthday(root_person, age=0)

print([person.age for person in person_list])

#calibrate the birthyear
min_generated_age = min([person.age for person in person_list])
min_age = 1800
for person in person_list:
    person.age = max(2000 - (person.age - min_generated_age), min_age)

age_list = np.asarray([person.age for person in person_list])
plt.hist(age_list)
plt.show()

age_mother = []
age_father = []
age_diff_couple = []
for edge, edge_type in zip(edge_list, edge_type_list):
    if edge_type == "mother":
        age_mother.append(person_list[edge[1]].age - person_list[edge[0]].age)
    elif edge_type == "father":
        age_father.append(person_list[edge[1]].age - person_list[edge[0]].age)
    elif edge_type == "husband":
        age_diff_couple.append(person_list[edge[1]].age - person_list[edge[0]].age)

plt.hist(age_mother)
plt.show()
plt.hist(age_father)
plt.show()
plt.hist(age_diff_couple)
plt.show()
```

```python
#2. birthplace
us_population_distribution = torch.load("us_population_distribution")
us_states_list = list(us_population_distribution.keys())
us_states_probs = np.asarray(list(us_population_distribution.values())).astype(np.float32); us_states_probs=us_states_probs/us_states_probs.sum()

p_parent = 0.9
for person in person_list:
    person.birthplace = None
    
def build_birthplace(person, birthplace=0):
    person.birthplace = birthplace
    
    if person.mother is not None:
        if person.mother.birthplace is None:
            mother_birthplace = birthplace if np.random.rand(1) < p_parent else np.random.choice(range(len(us_states_probs)), 1, p=us_states_probs)
            build_birthplace(person.mother, birthplace = mother_birthplace)

    if person.father is not None:
        if person.father.birthplace is None:
            father_birthplace = birthplace if np.random.rand(1) < p_parent else np.random.choice(range(len(us_states_probs)), 1, p=us_states_probs)
            build_birthplace(person.father, birthplace = father_birthplace)
            
    if person.children is not None:
        child_birthplace = birthplace if np.random.rand(1) < p_parent else np.random.choice(range(len(us_states_probs)), 1, p=us_states_probs)
        for child in person.children:
            if child.birthplace is None:
                build_birthplace(child, birthplace = child_birthplace)
    else:
        child_birthplace = birthplace if np.random.rand(1) < p_parent else np.random.choice(range(len(us_states_probs)), 1, p=us_states_probs)
        couple_birthplace = child_birthplace if np.random.rand(1) < p_parent else np.random.choice(range(len(us_states_probs)), 1, p=us_states_probs)
        
        if person.husband is not None:
            if person.husband.birthplace is None:
                build_birthplace(person.husband, birthplace=couple_birthplace)


        if person.wife is not None:
            if person.wife.birthplace is None:
                build_birthplace(person.wife, birthplace=couple_birthplace)

np.random.seed(4)
init_birthplace = np.random.choice(range(len(us_states_probs)), 1, p=us_states_probs)
build_birthplace(root_person, birthplace=init_birthplace)

birthplace_list = [person.birthplace[0] for person in person_list]
print(birthplace_list)

birthplace_dict = {}
for i in range(len(us_states_list)):
    birthplace_dict[i] = 0
    
for birthplace in birthplace_list:
    birthplace_dict[birthplace] += 1

plt.figure(figsize=(20, 5))
plt.bar(us_states_list, birthplace_dict.values())
plt.xticks(range(len(us_states_list)), us_states_list, rotation=45, fontsize=20)
plt.show()
```
